```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
games = pd.read_csv("data.csv", names = ["game", "genre", "platform", "release_date", "series", "developer", "tier"], skiprows=1)
games.head()
```

# Genre

```python
print(games.genre.value_counts())

counts = games.genre.value_counts()
other = 0
genre_remove = []
for k,v in counts.iteritems():
    if v < 4:
        other += v
        genre_remove.append(k)
        
counts['Other'] = other
for k in genre_remove:
    del counts[k]
genre_counts = counts.sort_values(ascending=False)
```

```python
# plt.figure(figsize=(20,10))
# plt.pie(genre_counts.values, labels=genre_counts.keys(), autopct='%1.1f%%', startangle=90)
# plt.axis('equal')
# plt.show()
```

```python
sns.set(style="whitegrid")
sns.set(rc={'figure.figsize':(20,10)})
sns.set(font_scale = 1.4)

ax = sns.barplot(x=genre_counts.keys(), y=genre_counts.values)
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
ax.set(title='Games per Genre', xlabel='Count of Games', ylabel='Genre')
```

# Series

```python
print(games.series.value_counts())
counts = games.series.value_counts()
other = 0
series_remove = []
for k,v in counts.iteritems():
    if v == 2:
        other += v
        series_remove.append(k)
    elif v < 2:
        series_remove.append(k)
        
counts['Other'] = other
for k in series_remove:
    del counts[k]
    
series_counts = counts.sort_values(ascending=False)
```

```python
# plt.figure(figsize=(20,10))
# plt.pie(series_counts.values, labels=series_counts.keys(), autopct='%1.1f%%', startangle=90)
# plt.axis('equal')
# plt.show()
```

```python
sns.set(font_scale = 1.4)

ax = sns.barplot(x=series_counts.keys(), y=series_counts.values)
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
ax.set(title='Games per Series', xlabel='Series', ylabel='Count of Games')
```

# Developer

```python
print(games.developer.value_counts())
counts = games.developer.value_counts()
other = 0
developer_remove = []
for k,v in counts.iteritems():
    if v <= 2:
        other += v
        developer_remove.append(k)
        
counts['Other'] = other
for k in developer_remove:
    del counts[k]
    
developer_counts = counts.sort_values(ascending=False)
```

# Average Tier per Genre

```python
genre = games.groupby('genre', as_index=False)['tier'].agg(['count', np.mean,np.std, np.median, np.min, np.max])
genre = genre[genre['count']>2]
genre = genre.sort_values(by='mean', ascending=False)
genre
```

# Average Tier per Platform

```python
platform = games.groupby('platform', as_index=False)['tier'].agg(['count', np.mean,np.std, np.median, np.min, np.max])
platform = platform.sort_values(by='mean', ascending=False)
platform
```
